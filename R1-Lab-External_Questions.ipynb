## About the dataset

This is perhaps the best known database to be found in the pattern recognition literature. The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant.  One class is linearly separable from the other 2; the latter are NOT linearly separable from each other.

Attribute Information:
   1. sepal length in cm
   2. sepal width in cm
   3. petal length in cm
   4. petal width in cm
   5. class: 
      -- Iris Setosa
      -- Iris Versicolour
      -- Iris Virginica

### Find the range of all the columns in the dataset

```python
print('SepalLength Range: ', iris['SepalLength'].max() - iris['SepalLength'].min())
print('SepalWidth Range: ', iris['SepalWidth'].max() - iris['SepalWidth'].min())
print('PetalLength Range: ', iris['PetalLength'].max() - iris['PetalLength'].min())
print('PetalWidth Range: ', iris['PetalWidth'].max() - iris['PetalWidth'].min())
print('SepalExtra Range: ', iris['SepalExtra'].max() - iris['SepalExtra'].min())
```
