```python
# we will work on our MovieLens dataset using the "Pandas" package.
# Pandas makes working with Tabular data very easy as we will see

import pandas as pd

def Goal_diff_count (a, b):
    return a - b

teams_df = pd.read_csv('Laliga.csv')
```

```python
# Cannot assign inplace String. So, changing it to numpy and setting 0
# teams_df[teams_df == "-"] = str("0")
# teams_numpy = teams_df.to_numpy()
# teams_numpy[teams_numpy =='-'] = 0

# Problem 1
#Changing - to 0 in dataframe
# for index, row in teams_df.iterrows():
#    row[row == '-'] ='0'
#    print(row)
teams_df = teams_df.replace(to_replace='-', value='0')
teams_df
```

```python
# Converting Strings datatype to integer for the required columns
start_debut = list()
points = list()
goalsFor = list()
goalsAgainst = list()
gamesPlayed = list()
gamesWon = list()

for index, row in teams_df.iterrows():
    start_debut.append(int(row['Debut'][0:4]))
    points.append(int(row['Points']))
    goalsFor.append(int(row['GoalsFor']))
    goalsAgainst.append(int(row['GoalsAgainst']))
    gamesPlayed.append(int(row['GamesPlayed']))
    gamesWon.append(int(row['GamesWon']))

# Adding StartDebut new column
teams_df['StartDebut'] = start_debut
# Change Points to interger
teams_df['Points'] = points
teams_df['GoalsFor'] = goalsFor
teams_df['GoalsAgainst'] = goalsAgainst
teams_df['GamesPlayed'] = gamesPlayed
teams_df['GamesWon'] = gamesWon
```

```python
# Problem 2
#Creating a filter for StartDebut year
filter = teams_df.apply(lambda row: (row['StartDebut'] > 1930) & (row['StartDebut'] < 1980), axis=1)
filter
#Selecting team based on filter
required_teams = teams_df[filter]
print('Team started debut between 1930 and 1980:')
required_teams.loc[:, ['Team', 'StartDebut']]
```

```python
# Problem 3
teams_df.sort_values(by='Points', ascending=False)
print('Teams with maximum Points:')
teams_df.loc[0:4, ['Team', 'Points']]
```

```python
# Problem 4
# Goal difference
goalDiff = Goal_diff_count(teams_df['GoalsFor'], teams_df['GoalsAgainst']);
teams_df['GoalsDiff'] = goalDiff

print('Team with maximum Goal difference:')
print(teams_df.loc[teams_df['GoalsDiff'].idxmax(), ['Team', 'GoalsDiff']])

print('Team with minimum Goal difference:')
print(teams_df.loc[teams_df['GoalsDiff'].idxmin(), ['Team', 'GoalsDiff']])
```

```python
# Problem 5

#percentWinning = (gamesWon / gamesPlayed)*100
filter = teams_df.apply(lambda row: 0 if row.loc['GamesPlayed'] == 0 else (row.loc['GamesWon'] / row.loc['GamesPlayed'])*100, axis=1)

# Adding new Column of "Percentage of Winning"
teams_df['Winning Percentage'] = filter
teams_df.sort_values(by="Winning Percentage", ascending=False)
teams_df.loc[0:4, ['Team', 'Winning Percentage']]
```

```python
# Problem 6
# This shows all Columns which can be summed:
#teams_df.groupby('BestPosition').sum()

# This shows only selected Columns against which summation is made
print('Teams aggregated by Best Position')
teams_df.groupby('BestPosition').agg({"Points": "sum"})
```
